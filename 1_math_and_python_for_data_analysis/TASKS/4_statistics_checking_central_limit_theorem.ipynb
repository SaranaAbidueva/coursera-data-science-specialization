# Бета-распределение

https://ru.wikipedia.org/wiki/%D0%91%D0%B5%D1%82%D0%B0-%D1%80%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5

параметры альфа=2, бета=2

```python
import scipy.stats as sts
import matplotlib.pyplot as plt
import numpy as np
```

генерирую выборку объема 1000, строю гистограмму и график теоретической плотности распределения

```python
a = 2  # выставляю параметры бета-распределения
b = 2

beta_f = sts.beta(a,b)
x = np.linspace(0, 1, 1000)
sample = sts.beta.rvs(a, b, size=1000)  # выборка 1000 значений
```

```python
plt.plot(x, beta_f.pdf(x),'r-', label='теоретическая плотность распределения')
plt.hist(sample, density=True)

plt.legend(loc='upper right')
plt.xlabel('$x$')
plt.ylabel('pdf/кол-во экземпляров')
plt.ylim([0, 2])
plt.plot
```

воспользуемся формулами теоретических матожидания и дисперсии из википедии

$E[x]=\frac{\alpha}{\alpha+\beta}$

$D[x]=\frac{\alpha \beta}{(\alpha+\beta)^2(\alpha+\beta+1)} $

```python
E = a / (a + b) 
D = a * b /((a + b)**2 *(a + b + 1)) 
print('E[x]=', E, ' D[x]=', D)
```

# Оценим распределение выборочного среднего случайной величины при разных объёмах выборок

# n=10, 50, 100

```python
for n in 10, 50, 100:
    sample_mean=[None]*1000 #массив выборочных средних
    for i in range(1000): #цикл генерации случайных выборок количества n из функции
        sample_n = beta_f.rvs(n) #создаю выборку
        sample_mean[i]=sum(sample_n)/n #подсчет выборочного среднего
    
    #определим параметры нормального распределения    
    sigma_n=(D/n)**(1/2)
    norm_n = sts.norm(loc=E, scale=sigma_n) # нормальное распределение
    
    plt.plot(x, norm_n.pdf(x), label='PDF(n=%s)' % n)
    plt.hist(sample_mean, alpha=0.5, density=True, label='sample mean n=%s' % n)
    plt.ylabel('number of samples/pdf')
    plt.xlabel('$x$')
    plt.legend(loc='best')
```

Таким образом, при увеличении объема выборки распределение выборочного среднего коррелирует в соответствии с центральной предельной теоремой. Также увеличивается точность аппроксимации выборочного среднего. Сужаются доверительные интервалы, а плотность вероятности увеличивается
