Задача 2: аппроксимация функции

Рассмотрим функцию на отрезке [1, 15]:

f(x) = sin(x / 5) * exp(x / 10) + 5 * exp(-x / 2)

```python
import matplotlib.pyplot as plt
import numpy as np
y = lambda x: np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
A = np.matrix([[1, 1], [1, 15]])
b = [y(1), y(15)]
sol = np.linalg.solve(A, b)
x = np.linspace(0, 16, 100)
y1 = lambda x: sol[0]+sol[1]*x
plt.plot(x, y(x))
plt.plot(x, y1(x))
```

```python
A = np.matrix([[1, 1, 1],[1, 8, 64],[1, 15, 225]])
b = [y(1), y(8), y(15)]
sol = np.linalg.solve(A, b)
x = np.linspace(0, 16, 100)
y1 = lambda x: sol[0]+sol[1]*x+sol[2]*x**2
plt.plot(x, y(x))
plt.plot(x, y1(x))
```

```python
A = np.matrix([[1, 1, 1, 1],[1, 4, 16, 4**3],[1, 10, 100, 1000], [1, 15, 225, 15**3]])
b = [y(1), y(4), y(10), y(15)]
sol = np.linalg.solve(A, b)
x = np.linspace(0, 16, 100)
y1 = lambda x: sol[0]+sol[1]*x+sol[2]*x**2+sol[3]*x**3
plt.plot(x, y(x))
plt.plot(x, y1(x))
```

```python
sol = np.around(sol, 2)
f = open('submission2.txt', 'w')
string = str(sol[0]) + ' ' + str(sol[1]) + ' ' + str(sol[2]) +' '+ str(sol[3])
f.write(string)
f.close()
```
